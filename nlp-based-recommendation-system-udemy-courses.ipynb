```python
# Identify outliers using the IQR method
outliers_count = {}

for column in numerical_columns:
    Q1 = df[column].quantile(0.25)  # First quartile (25th percentile)
    Q3 = df[column].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile Range

    lower_bound = Q1 - 1.5 * IQR  # Lower boundary
    upper_bound = Q3 + 1.5 * IQR  # Upper boundary

    # Filter outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    outliers_count[column] = outliers.shape[0]  # Count the number of outliers

# Print the total number of outliers for each column
for column, count in outliers_count.items():
    print(f"Column {column} has {count} outliers.")
```

# Recommendation System

# TF-IDF 

Since I don't have any user behavior data, I'm going to develop a recommendation system based on categories and course features. A content-based system can be created with variables such as category, rating, and instructional_level. However, content-based systems are generally limited to structured data, which can reduce the diversity of recommendations. Moreover, broad and generalizing features like 'category' may not capture the specific details of courses sufficiently. For example, there may be significant differences between two courses in the "Personal Development" category, but these differences cannot be distinguished by the system.

By using NLP, a system can be developed that can learn the content features of courses in more depth by capturing these differences. This provides more meaningful and personalized recommendations. By perceiving courses with similar meanings through textual similarities, a richer recommendation pool can be created. In this case, the roadmap I will follow when developing the recommendation system is:

1- Analyzing the texts of the courses with NLP and creating a recommendation network based on semantic similarities.

2- Supporting this network by filtering it with structured features such as instructional_level or category.

```python
# Combine text columns for NLP analysis
df['combined_text'] = df['headline'] + ' ' + df['objectives'] + ' ' + df['curriculum']

# Text cleaning function
def clean_text(text):
    text = re.sub(r'\W+', ' ', text)  # Remove special characters
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = text.lower()  # Convert to lowercase
    return text

df['cleaned_text'] = df['combined_text'].apply(clean_text)
df
```

NLP models perform better on clean and structured text. Special characters, capitalization differences, and other noise can affect model performance.

df['combined_text'] = df['headline'] + ' ' + df['objectives'] + ' ' + df['curriculum']: This line combines the text from the 'headline', 'objectives', and 'curriculum' columns into a single 'combined_text' column, separated by spaces.

clean_text(text): This function cleans the text by performing the following operations:

re.sub(r'\W+', ' ', text): Replaces special characters in the text with spaces. For example, punctuation marks, numbers, etc.
text.lower(): Converts the text to lowercase. This eliminates case sensitivity during text analysis.
df['cleaned_text'] = df['combined_text'].apply(clean_text): This line applies the clean_text function to each text in the 'combined_text' column and assigns the cleaned text to a new 'cleaned_text' column.

```python
# Convert cleaned text to TF-IDF vectors
tfidf = TfidfVectorizer(max_features=5000)  # Adjust max_features as needed
tfidf_matrix = tfidf.fit_transform(df['cleaned_text'])

# DataFrame of TF-IDF features
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
```

This code converts textual data into numerical features that can be processed by machine learning algorithms. These numerical representations capture the semantics and similarity of texts, making them applicable to a wide range of natural language processing tasks.

**Why I Started with TF-IDF and Transitioned to BERT**

In the initial stages of the project, I used TF-IDF (Term Frequency-Inverse Document Frequency) to extract features from course descriptions. TF-IDF is a well-known method for text analysis, which helped me identify important words in the documents and reduced the weight of common terms in the dataset. The main reasons I used TF-IDF at this stage were:


Efficiency: TF-IDF has a lower computational cost compared to deep learning models and allows for quick processing of large datasets.

Simplicity: It offers an effective method for capturing word importance without considering context. This was sufficient for the initial exploration phase.

Initial Performance: TF-IDF provided a good baseline for the performance of the recommendation system, helping me develop an early understanding of relationships within the data.


**Why I Transitioned to BERT**

Although I was able to build a simple recommendation system with TF-IDF, it had limitations in capturing context and semantic relationships. BERT (Bidirectional Encoder Representations from Transformers) processes words both left-to-right and right-to-left, considering the full context of the text. Therefore, I decided to switch to BERT because:


Contextual Understanding: While TF-IDF treats words in isolation, BERT captures the contextual meaning of words. This provides a critical advantage in cases where the meaning depends on the entire document, especially in descriptive texts.

Advanced Semantic Relationships: BERT can better grasp semantic relationships between words and phrases. For instance, where TF-IDF might fall short in understanding the relationship between "beginner" and "intermediate" levels, BERT can capture this relationship more accurately.

Better Performance: With its deep learning structure, I expect BERT to provide higher accuracy and more accurate embeddings based on context, leading to improved performance for the recommendation system.

Scalability: While BERT requires more computational power and time, its ability to improve recommendation accuracy and understand the complex nuances of text will justify the initial investment in the long term.

By transitioning to BERT, I aim to provide more personalized and contextually relevant course recommendations.

```python
# Load the BERT model and tokenizer
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
```

```python
# Check if GPU is available and set the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load the DistilBERT model and tokenizer to the selected device
model_name = 'distilbert-base-uncased'
tokenizer = DistilBertTokenizer.from_pretrained(model_name)
model = DistilBertModel.from_pretrained(model_name).to(device)

def get_distilbert_embeddings(texts, model, tokenizer, max_length=512, batch_size=32, device='cpu'):
    """
    Generate DistilBERT embeddings for a list of texts using GPU if available.
    """
    embeddings = []
    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i + batch_size]
        # Move inputs to GPU
        inputs = tokenizer(
            batch_texts, 
            max_length=max_length, 
            padding=True, 
            truncation=True, 
            return_tensors='pt'
        ).to(device)
        
        with torch.no_grad():
            # Compute embeddings on GPU
            outputs = model(**inputs)
        # CLS token embedding (first vector of each sequence), move back to CPU
        batch_embeddings = outputs.last_hidden_state[:, 0, :].cpu().numpy()
        embeddings.extend(batch_embeddings)
    
    return embeddings  # Return a list of embeddings


embeddings = get_distilbert_embeddings(
    df['cleaned_text'].tolist(), 
    model,
    tokenizer,
    device=device
)

df['bert_embeddings'] = embeddings

df
```
